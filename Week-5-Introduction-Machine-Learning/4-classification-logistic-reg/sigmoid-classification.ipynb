```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sigmoid Function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
# Generate value
z = np.linspace(-10, 10, 100)
sigmoid_values = sigmoid(z)
```

```python
# Plot
plt.plot(z, sigmoid_values)
plt.title("Sigmoid Function")
plt.xlabel("z")
plt.ylabel("Sigma(z)")
plt.grid()
plt.show()
```
